# BlogFeedback Prediction - Using Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Read training dataset
train = pd.read_csv("blogData_train.csv", header=None)
train.head(10)
```

```python
#Read test dataset
test = pd.read_csv("blogData_test-2012.03.31.01_00.csv", header=None)
test.head(10)
```

```python
#Divide training dataset in x and y
X_train = train.iloc[:,0:280]
y_train = train.iloc[:,-1]
```

```python
#Build linear regression model
lin_model = LinearRegression()
```

```python
#Train model on training dataset
lin_model.fit(X_train, y_train)
```

```python
#Predict using model

y_train_predict = lin_model.predict(X_train)

#Calculate mean squared error and root mean square error
mse = mean_squared_error(y_train, y_train_predict)
rmse = np.sqrt(mse)

print("The model performance for training dataset:\n")
print("MSE :", mse)
print("RMSE :",rmse)
```

```python
#Divide test dataset in x and y
X_test = test.iloc[:,0:280]
y_test = test.iloc[:,-1]
```

```python
#Predict target on test data using model

y_test_predict = lin_model.predict(X_test)

mse = mean_squared_error(y_test, y_test_predict)
rmse = np.sqrt(mse)

print("The model performance for testing dataset:\n")
print("MSE :", mse)
print("RMSE :",rmse)
```
